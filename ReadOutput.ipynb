```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sequential_time = 116.875
```

```python
test1 = pd.read_csv('src/test1_out.csv')
test2 = pd.read_csv('src/test2_out.csv')
test3 = pd.read_csv('src/test3_out.csv')

# test1['PS'] = sequential_time / test1['elapsed_time']
# test1['PE'] = test1['PS'] / test1['number_of_processes']

# test2['PS'] = sequential_time / test2['elapsed_time']
# test2['PE'] = test2['PS'] / test2['number_of_processes']

test3['PS'] = sequential_time / test3['elapsed_time']
test3['PE'] = test3['PS'] / test3['number_of_processes']
```

```python
# test1.to_csv('src/test1_out.csv',index=False)
# test2.to_csv('src/test2_out.csv',index=False)
test3.to_csv('src/test3_out.csv',index=False)
```

```python


t1 = sns.lineplot(x = test1['number_of_processes'],
               y = test1['PS'])
t1.set_xlabel('Number of Processes')
t1.set_ylabel('Parallel Speedup')
#t1.set_title('Parallel Speedup of MPI 0 using 2,4,8 processes')
plt.savefig('img/test1_PS.png')
```

```python
t2 = sns.lineplot(x = test1['number_of_processes'],
               y = test1['PE'])
t2.set_xlabel('Number of Processes')
t2.set_ylabel('Parallel Efficiency')
#t2.set_title('Parallel Efficiency of MPI 0 using 2,4,8 processes')
plt.savefig('img/test1_PE.png')
```
